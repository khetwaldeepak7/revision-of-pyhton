##   --- >>>>>>>>... WEB SCRAPING USING REQUESTS MODULE <<<<<<<<--------------

```python
import requests
import re
import json
import pandas as pd
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36' }
r=requests.get("https://www.flipkart.com/search?q=best+laptop&as=on&as-show=on&otracker=AS_Query_HistoryAutoSuggest_1_6_na_na_na&otracker1=AS_Query_HistoryAutoSuggest_1_6_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=best+laptop&requestId=3d069906-547e-4ca0-ac61-497dff07364b&as-backfill=on",headers=headers).text     #by using this it will convert it into the text format
```

```python
print(r)  # esmai sabh data hai,naya variable create hoga,and usmai sabh essential data stored hoga
```

```python
alldata=re.search('<div class="_1UoZlX">(.*?)<div class="_2zg3yZ">',r).group(1) # inside this i have stored only the essetial data and unwanted are remove
```

```python
#print(alldata)
# split  sey ek product ko dusre product sey split karenge,,split by mean ,split ke andar jo value dia hai ,use split hoga

# alldata mai sab data aa gya,abh har ek product ko ek sudre sey alag karenge ,and data we got in the format of string,so abh hum 
# string ko split karenge, alldata.split(value).mean alldatako split karna hai (bracket ke andar values)
names=[]
urls=[]
pri=[]
for unqiue in alldata.split('<a class="_31qSD5"')[1:]:  # here i split everypaticular product ,jaise pehle hp hai n dusra alag,so here i have
    #print(unqiue)
    #urlofproduct= re.search('href=(.*?)\/',unqiue).group(1)
    
    #print(urlofproduct)
    url=re.search('href="(.*?);',unqiue,re.S).group(1)
    page_urls="https://www.flipkart.com"+url
    #print('https://www.flipkart.com'+url)
    name=re.search('<div class="_3wU53n">(.*?)<\/div>',unqiue,re.S).group(1)
    price=re.search('<div class="_1vC4OE _2rQ-NK"(.*?)<',unqiue).group(1)
    prices=re.sub('>','',price)  # by using subsitution i will subsituate the value ,and 
    # sub ->take 3 argument, 1)kisko subsitute karna hai ,2) kisse subsituate karna hai, 3)kismai sey sey subsituate karna hai
    names.append(name)  # this i wil have to append in a list,so i can used it
    urls.append(page_urls)
    pri.append(prices)
    
```

```python
#url=re.search('href="(.*?);',unqiue,re.S).group(1)  # as i had split ,and we know with the help of split it will store data in a list format
# so first iterate mai ek list ayega,and usmai fir search karenge,url ,name,image and price
```

```python
for i in pri:  # iterating one by one url
#without subsituting ->>   i=re.sub('₹','',i)  # this is how also we can subsitute
    print("actual prise of product is: "+i)   # it will print ₹ sign as well ,before subsituting the values
   
```

```python
df=pd.DataFrame(
{
    'name':names,
    'url':urls,
    'price':pri
})
```
