```python
import requests
import re
```

```python
r=requests.get("https://timesofindia.indiatimes.com/india/coronavirus-in-india-live-updates-covid-19-22-nov-2020/liveblog/79346759.cms").text
```

```python
#print(r)
```

```python
header=re.search("<h1>(.*?)</h1>",r).group(1)
```

```python
print(header)
```

```python
for date in r.split('class="xOe7d"')[1:]:
    tarik=re.search(">(.*?)</time>",date).group(1)
    print(tarik)
```

```python
image=re.search('<div class="_3qAmt">.*?src="(.*?)"',r).group(1)
```

```python
print(image)
```

```python
 img=requests.get(image).content     # for requests.get is a method,
```

```python
# now i wil write it in the file
with open("image1.png",'wb') as F:
    F.write(img)
    
```
